```python
from scipy.signal import find_peaks

def extract_pd_features(csv_file):
    df = pd.read_csv(csv_file, header=[0,1,2])
    df.columns = pd.MultiIndex.from_tuples([(bp, coord) for (_, bp, coord) in df.columns])

    # Extract thumb & index
    thumb_x = df['thumb','x']
    thumb_y = df['thumb','y']
    thumb_p = df['thumb','likelihood']

    index_x = df['indexfinger','x']
    index_y = df['indexfinger','y']
    index_p = df['indexfinger','likelihood']

    # Confidence filter
    mask = (thumb_p > 0.8) & (index_p > 0.8)
    thumb_x = thumb_x[mask]
    thumb_y = thumb_y[mask]
    index_x = index_x[mask]
    index_y = index_y[mask]

    # Distance signal
    dist = np.sqrt((thumb_x-index_x)**2 + (thumb_y-index_y)**2)

    # Detect taps (valleys)
    inv = -dist.values
    peaks,_ = find_peaks(inv, distance=15, prominence=10)

    if len(peaks) < 5:
        return None  # bad trial

    # Biomarkers
    iti = np.diff(peaks)/240
    amp = dist.values[peaks]
    fatigue = np.polyfit(np.arange(len(amp)), amp, 1)[0]

    return {
        "tap_rate": len(peaks)/(len(dist)/240),
        "mean_iti": np.mean(iti),
        "std_iti": np.std(iti),
        "mean_amp": np.mean(amp),
        "std_amp": np.std(amp),
        "fatigue": fatigue
    }
```

```python
import numpy as np
import pandas as pd
from scipy.signal import find_peaks

def extract_temporal_fatigue_features(csv_path):
    """
    Temporal fatigue extraction with Option A:
    inverted distance so negative slope = motor deterioration
    """

    df = pd.read_csv(csv_path, header=[0,1,2])

    # --- Extract thumb & index finger ---
    thumb = df.xs("thumb", level=1, axis=1)
    index = df.xs("indexfinger", level=1, axis=1)

    thumb_x = thumb.xs("x", level=1, axis=1).values.squeeze()
    thumb_y = thumb.xs("y", level=1, axis=1).values.squeeze()
    index_x = index.xs("x", level=1, axis=1).values.squeeze()
    index_y = index.xs("y", level=1, axis=1).values.squeeze()

    # --- OPTION A: Inverted distance ---
    distance = -np.sqrt((thumb_x - index_x)**2 + (thumb_y - index_y)**2)
    distance = np.asarray(distance).squeeze()

    # --- Detect taps ---
    inv_dist = -distance  # revert sign ONLY for peak detection
    peaks, _ = find_peaks(inv_dist, distance=10, prominence=5)

    if len(peaks) < 5:
        return None

    # --- Time-normalized segmentation ---
    total_frames = len(distance)
    early_end = total_frames // 3
    mid_end   = 2 * total_frames // 3

    early_taps = peaks[peaks < early_end]
    mid_taps   = peaks[(peaks >= early_end) & (peaks < mid_end)]
    late_taps  = peaks[peaks >= mid_end]

    if min(len(early_taps), len(mid_taps), len(late_taps)) < 2:
        return None

    # --- Amplitude fatigue (hypokinesia) ---
    amp_early = np.mean(distance[early_taps])
    amp_mid   = np.mean(distance[mid_taps])
    amp_late  = np.mean(distance[late_taps])

    amp_fatigue_slope = np.polyfit(
        [0, 1, 2],
        [amp_early, amp_mid, amp_late],
        1
    )[0]

    # --- Inter-tap velocity fatigue (bradykinesia) ---
    intervals = np.diff(peaks)
    amplitudes = distance[peaks[:-1]]
    intertap_velocity = amplitudes / intervals

    vel_early = np.mean(intertap_velocity[: len(intertap_velocity)//3])
    vel_mid   = np.mean(intertap_velocity[len(intertap_velocity)//3 : 2*len(intertap_velocity)//3])
    vel_late  = np.mean(intertap_velocity[2*len(intertap_velocity)//3 :])

    vel_fatigue_slope = np.polyfit(
        [0, 1, 2],
        [vel_early, vel_mid, vel_late],
        1
    )[0]

    return {
        "amp_fatigue_slope": amp_fatigue_slope,
        "vel_fatigue_slope": vel_fatigue_slope,
        "num_taps": len(peaks)
    }
```

```python
def inter_tap_velocity(peaks, distance):
    if len(peaks) < 2:
        return np.nan
    intervals = np.diff(peaks)
    amplitudes = distance[peaks[:-1]]
    return np.mean(amplitudes / intervals)
```

```python
import numpy as np

pd_vals = subject_df[subject_df["label"] == 1]["vel_fatigue_slope"]
hc_vals = subject_df[subject_df["label"] == 0]["vel_fatigue_slope"]

def cohens_d(x, y):
    nx, ny = len(x), len(y)
    pooled_std = np.sqrt(
        ((nx-1)*np.var(x, ddof=1) + (ny-1)*np.var(y, ddof=1)) / (nx+ny-2)
    )
    return (np.mean(x) - np.mean(y)) / pooled_std

cohens_d(pd_vals, hc_vals)
```
